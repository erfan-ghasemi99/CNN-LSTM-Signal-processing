```python
import mne
import numpy as np
import pandas as pd
import os
```

```python
# .edf file dir
edf_path_list = ['h01.edf','h02.edf','h03.edf','h04.edf','h05.edf','h06.edf','h07.edf','h08.edf','h09.edf','h10.edf','h11.edf','h12.edf','h13.edf','h14.edf','s01.edf',
's01.edf','s02.edf','s03.edf','s04.edf','s05.edf','s06.edf','s07.edf','s08.edf','s09.edf','s10.edf','s11.edf','s12.edf','s13.edf','s14.edf']
```

```python
# Load EEG Signal
raw = mne.io.read_raw_edf(f'data/{edf_path_list[15]}', preload=True)
eeg_channels = ['Fp1', 'Fp2', 'F7', 'F3', 'Fz', 'F4', 'F8',
                'T3', 'C3', 'Cz', 'C4', 'T4',
                'T5', 'P3', 'Pz', 'P4', 'T6', 'O1', 'O2']
raw.pick_channels(eeg_channels)
```

```python
# Make Output Folder
os.makedirs("eeg_csv_frames", exist_ok=True)
```

```python
for sample in range(len(edf_path_list)):
# Load EEG Signal
    raw = mne.io.read_raw_edf(f'data/{edf_path_list[sample]}', preload=True)
    eeg_channels = ['Fp1', 'Fp2', 'F7', 'F3', 'Fz', 'F4', 'F8',
                'T3', 'C3', 'Cz', 'C4', 'T4',
                'T5', 'P3', 'Pz', 'P4', 'T6', 'O1', 'O2']
    raw.pick_channels(eeg_channels)

    data, _ = raw.get_data(return_times=True)
    sfreq = int(raw.info['sfreq'])  # باید 250 باشه
    frame_size = 25 * sfreq  # 25 ثانیه = 6250 نمونه

    # Frame
    n_frames = data.shape[1] // frame_size

    os.makedirs(f"eeg_csv_frames/{edf_path_list[sample][:3]}", exist_ok=True)

    for i in range(n_frames):
        frame = data[:, i * frame_size : (i + 1) * frame_size].T  # shape: (6250, 19)
        df = pd.DataFrame(frame, columns=eeg_channels)
        df.to_csv(f"eeg_csv_frames/{edf_path_list[sample][:3]}/frame_{i:02}.csv", index=False)

    print(f"Saved (({n_frames})) from  {edf_path_list[sample]}  File.  ✅")
```
